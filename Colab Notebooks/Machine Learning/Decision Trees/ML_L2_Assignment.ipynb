```python
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.ensemble import VotingClassifier
SEED=1
```

```python
X=titanic_data.drop(['Survived','Name','Sex','Ticket','Embarked'],axis=1)
y=titanic_data.Survived
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
lr=LogisticRegression(random_state=SEED)
knn=KNN()
dt=DecisionTreeClassifier(random_state=SEED)
classifiers=[('Logistic Regression',lr),('K Nearest Neighbors',knn),('Classification Tree',dt)]
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
for clf_name,clf in classifiers:
  clf.fit(X_train,y_train)
  y_pred=clf.predict(X_test)
  print('{:s}:{:2f}'.format(clf_name,accuracy_score(y_test,y_pred)))
```

```python
vc=VotingClassifier(estimators=classifiers)
vc.fit(X_train,y_train)
y_pred=vc.predict(X_test)
print("Voting Classifier:",round(accuracy_score(y_test,y_pred),2))
```

```python
from sklearn.ensemble import BaggingClassifier
```

```python
dt = DecisionTreeClassifier(max_depth=4, min_samples_leaf=0.16, random_state=SEED)
bc=BaggingClassifier(base_estimator=dt,n_estimators=300,n_jobs=-1)
```

```python
bc.fit(X_train,y_train)
```

```python
y_pred=bc.predict(X_test)
```

```python
accuracy=accuracy_score(y_test,y_pred)
```

```python
print("Accuracy:",round(accuracy,3))
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf=RandomForestRegressor(n_estimators=400,min_samples_leaf=0.2,random_state=SEED)
```

```python
rf.fit(X_train,y_train)
y_pred=rf.predict(X_test)
mse=MSE(y_test,y_pred)
print("Accuracy:",round(mse**0.5))
```

```python
importance_rf=pd.Series(rf.feature_importances_,index=X.columns)
```

```python
sorted_importance_rf=importance_rf.sort_values()
```

```python
sorted_importance_rf.plot(kind='barh')
plt.show()
```

```python
X=titanic_data[['Fare','Pclass']]
y=titanic_data.Survived
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import roc_auc_score
```

```python
X_train, X_test, y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
dt=DecisionTreeClassifier(max_depth=2,random_state=SEED)
ada=AdaBoostClassifier(base_estimator=dt,n_estimators=200)
ada.fit(X_train,y_train)
y_pred=ada.predict_proba(X_test)[:,1]
```

```python
ada_score=roc_auc_score(y_test,y_pred)
```

```python
print("ROC AUC Score",round(ada_score,2))
```

```python
from sklearn.metrics import roc_curve
fper,tper,thresholds=roc_curve(y_test,y_pred)
```

```python
plt.plot(fper,tper)
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False Postive Rate')
plt.ylabel('True Positive Rate')
plt.title('Adacoost ROC curve')
plt.show()
```
