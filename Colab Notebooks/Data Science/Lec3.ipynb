```python
import numpy as np
import pandas as  pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
height=[3,4,12,5,6]
bars=['A','B','C','D','E']
plt.bar(bars,height)
plt.xticks(bars)#Xaxis name
plt.show()
```

```python
height=[3,4,12,5,6]
bars=['A','B','C','D','E']
plt.barh(bars,height)
plt.yticks(bars)#Xaxis name
plt.show()
```

```python
plt.bar(bars,height,color=(0.4,0.4,0.2,0.8)) #(R,G,B,Opacity)
plt.xticks(bars)
plt.show()
```

```python
plt.bar(bars,height,color=['black','red','green','blue','cyan']) #(R,G,B,Opacity)
plt.xticks(bars)
plt.show()
```

```python
plt.bar(bars,height,color=(0.4,0.4,0.2,0.8),edgecolor='blue') #(R,G,B,Opacity)
plt.xticks(bars)
plt.show()
```

```python
plt.bar(bars,height,color=(0.4,0.4,0.2,0.8)) #(R,G,B,Opacity)
plt.xticks(bars)
plt.title('$Height_{bars}$')
plt.xlabel('Alphabet')
plt.ylabel('Number')
plt.ylim(0,15)
plt.show()
```

```python
df=pd.DataFrame({'year':[2016,2017,2018,2019],'T20':[22,32,12,25], 'ODI':[26,35,34,11]})
plt.plot('year','ODI',data=df)
plt.show()
```

```python
plt.plot('year','ODI',data=df,linewidth=7)
plt.plot('year','T20',data=df,linewidth=7)
plt.xticks(np.arange(min(df.year),max(df.year)+1,1),fontsize=20)
#plt.yticks((0,125,25),fontsize=20)
plt.xlabel('Year')
plt.ylabel('Points')
plt.legend(loc='upper right')
plt.grid(b=True, alpha=0.1)
plt.show()
```

```python
values=np.cumsum(np.random.randn(1000,1))
plt.plot(values)
```

```python
plt.plot('year','ODI',data=df,marker='o',markerfacecolor='blue',markersize=12,linewidth=7)
plt.plot('year','T20',data=df,linewidth=7)
plt.xticks(np.arange(min(df.year),max(df.year)+1,1),fontsize=20)
#plt.yticks((0,125,25),fontsize=20)
plt.xlabel('Year')
plt.ylabel('Points')
plt.legend(loc='upper right')
plt.grid(b=True, alpha=0.1)
plt.show()
```
