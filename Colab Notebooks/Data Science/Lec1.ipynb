**Linear Algebra**

```python
np.random.seed(4) #Gives same random no
a=np.random.random_sample((3,3))*5
b=np.exp(a)
print(b)
```

```python
#Rank,Determinent
det=np.linalg.det(b)
print("Determinent={}".format(det))
r=np.linalg.matrix_rank(b)
print("Rank={}".format(r))
t=np.trace(b)
print("Trace={}".format(t))
```

```python
#Eigenvalues/EigenVectors
eigen_val,eigen_vec=np.linalg.eig(b)
print("Eigenvalues={}".format(eigen_val))
print("Eigenvectors={}".format(eigen_vec))
```

```python
#SVD
u,v,vh=np.linalg.svd(b)
print("U={}".format(u))
print("V={}".format(v))
print("Singular values of b are={}".format(vh))
```
